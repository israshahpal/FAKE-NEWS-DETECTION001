```python
# --- 1. SETUP: Install Necessary Libraries ---
# No change here
!pip install transformers pandas numpy scikit-learn matplotlib seaborn
!pip install --quiet ipywidgets # For interactive report in Colabs

# --- 2. IMPORTS ---
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt # We will use Matplotlib for compact visualization
import seaborn as sns
from transformers import AutoTokenizer, AutoModelForSequenceClassification, pipeline
import torch
from IPython.display import display, Markdown, HTML

# --- 3. SIMULATE PAKISTAN DATASET (English Headlines) ---
# No change here, maintaining English data
data = {
    'news_text': [
        "Prime Minister announces immediate decrease in inflation; essential prices have fallen dramatically.", # Fake Claim
        "Pakistan secures $3 billion IMF loan to stabilize the national economy.", # Real Report
        "Electricity tariffs in Karachi will see a further increase, according to the Ministry of Energy.", # Real Report
        "The military has seized control and declared a state of emergency nationwide.", # Fake Rumour
        "The Pakistan Cricket Team has won the World Cup; fans celebrate in the streets.", # Fake Hype
        "New health initiative launched in Lahore focusing on child nutrition.", # Real Report
        "Petrol prices have suddenly been reduced to zero, says Ministry of Finance.", # Fake Claim
        "Chief Justice comments on pending cases and judicial reform proposals.", # Real Report
    ],
    'is_fake': [1, 0, 0, 1, 1, 0, 1, 0] # 1 = Fake, 0 = Real
}
df = pd.DataFrame(data)
df['label_name'] = df['is_fake'].apply(lambda x: 'FAKE NEWS ‚ùå' if x == 1 else 'REAL NEWS ‚úÖ')

display(Markdown("## üáµüá∞ Simulated Pakistan News Dataset (English)"))
display(df.head())
print("-" * 50)

# --- 4. LOAD PRE-TRAINED ENGLISH MODEL (BERT) ---
MODEL_NAME = "bert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
classifier = pipeline(
    "text-classification",
    model=MODEL_NAME,
    tokenizer=tokenizer,
    top_k=None
)
print(f"‚úÖ English Model Loaded: {MODEL_NAME}")
print("-" * 50)

# --- 5. REAL-TIME SIMULATION & COMPACT REPORT GENERATION FUNCTION ---

def generate_compact_report(input_text, classifier_pipeline):
    """Processes input, classifies, and generates a compact formatted report with Matplotlib visualization."""

    # 5.1 CLASSIFICATION
    try:
        results = classifier_pipeline(input_text)[0]
    except Exception as e:
        return display(Markdown(f"### üõë Error during Classification: {e}"))

    # Convert scores
    scores = {item['label']: item['score'] for item in results}
    fake_score = scores.get('LABEL_1', 0)
    real_score = scores.get('LABEL_0', 0)

    # Determine the final result
    if fake_score > real_score:
        result_label = "FAKE NEWS ‚ùå"
        result_color = "#E74C3C" # Red
        confidence = fake_score * 100
    else:
        result_label = "REAL NEWS ‚úÖ"
        result_color = "#2ECC71" # Green
        confidence = real_score * 100

    # 5.2 SOURCE TRACKING (SIMULATED)
    source_simulated = random.choice([
        "Social Media Forward (Unverified)",
        "Major National TV Channel Website",
        "Official Government Press Release",
        "Anonymous WhatsApp Group",
        "Verified News Agency"
    ])

    # 5.3 GENERATE COMPACT REPORT

    # Header & Input
    display(Markdown(f"""
    ---
    ### üì∞ Report: **"{input_text}"**
    """))

    # Summary Card (Minimizing padding/margins)
    html_card = f"""
    <div style="background-color: {result_color}; color: white; padding: 10px; border-radius: 5px; text-align: center; display: inline-block; margin: 5px 0;">
        <h3 style="margin: 0; font-size: 1.2em;">FINAL CLASSIFICATION: {result_label}</h3>
        <p style="margin: 0; font-size: 1em;">Confidence: **{confidence:.2f}%** | Source: {source_simulated}</p>
    </div>
    """
    display(HTML(html_card))

    # 5.4 VISUALIZATION (Compact Matplotlib Bar Chart)

    labels = ['REAL', 'FAKE']
    probabilities = [real_score, fake_score]
    colors = ['#2ECC71', '#E74C3C']

    plt.figure(figsize=(4, 1.5)) # Smaller figure size
    plt.barh(labels, probabilities, color=colors)

    # Add percentage labels inside bars
    for index, value in enumerate(probabilities):
        plt.text(value - 0.05, index, f'{value*100:.1f}%', va='center', color='white', fontweight='bold')

    plt.title('Confidence Breakdown', fontsize=8)
    plt.xlim(0, 1)
    plt.xticks([]) # Hide x-axis ticks
    plt.show()

    display(Markdown("\n")) # Small separation space


# --- 6. EXECUTE: TEST CASES (Simulating Real-Time Input) ---

print("## üöÄ Running Test Cases (Compact Output)")

# Test Case 1: Clear Fake News (High-stakes false claim)
input_1 = "The Ministry of Health announced that every Pakistani citizen will receive 50,000 rupees in free cash aid. Register quickly!"
generate_compact_report(input_1, classifier)

# Test Case 2: Real News (Official economic report)
input_2 = "The State Bank of Pakistan raised the key interest rate by 150 basis points to 22%, effective immediately."
generate_compact_report(input_2, classifier)

# Test Case 3: Ambiguous/Controversial News (Political event)
input_3 = "Opposition parties have accused the government of rigging the elections; the Election Commission has launched an investigation."
generate_compact_report(input_3, classifier)
```
